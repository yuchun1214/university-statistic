```python
import pandas as pd
import difflib
```

```python
data = pd.read_excel("./校庫學1﹍11003﹍1100316更新版.xls", sheet_name=2)
```

```python
config = pd.read_excel("./核對版第11003期_學1.正式學籍在學學生人數(3月、10月填報).xls", sheet_name=1)
```

```python
config_statistic = {}
for i in range(len(config.values)):
    if(config.at[i, "班次"] != 0):
        key = config.at[i, "單位"] + config.at[i, "學制班別"] + config.at[i, "班次"]
    else:
        key = config.at[i, "單位"] + config.at[i, "學制班別"]
    if(not key in config_statistic):
        config_statistic[key] = 0
    config_statistic[key] += config.at[i, "正式學籍之在學學生總人數男"] + config.at[i, "正式學籍之在學學生總人數女"] + config.at[i, "轉學生男"] + config.at[i, "轉學生女"]
```

```python
config = config.fillna(0)
```

```python
config_statistic
```

```python
maps = {}
keys = list(config_statistic.keys())
data.dept_name = data.dept_name.str.replace(r' ', '')
dept_names = list(data.dept_name)
for dept in dept_names:
    texts = difflib.get_close_matches(dept, keys)
    print(dept, texts)
    try:
        maps[dept] = texts[0]
    except:
        maps[dept] = ''
```

```python
import json
```

```python
text = json.dumps(maps, indent=4, ensure_ascii=False)
with open("dept_maps.json", "w")as file:
    file.write(text)
    file.close()
```

```python
for i in range(len(data.values)):
    data_dept_name = data.at[i, "dept_name"]
    config_dept_name = maps[data_dept_name]
    num1 = data.at[i, "合計"]
    num2 = config_statistic[config_dept_name]
    if(num2 != num1):
        print("Warning!!! config(%s, %d), data(%s, %d)" % (config_dept_name, num2, data_dept_name, num1))
```
